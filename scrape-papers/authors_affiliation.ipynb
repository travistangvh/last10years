```python
# Extract all Author(s) ID from the column "Author(s) ID" and store it into author_id_list
# The column "Author(s) ID" contains all the author IDs in the format "Author ID1; Author ID2; Author ID3"
# We need to split the string by ";" and then extract the author ID
# We will use the function "split" to split the string by ";"
# remember to remove any empty results or any leading or trailing spces

# Code here
author_id_list = []
for author_id in df['Author(s) ID'].drop_duplicates():
    author_id_list += author_id.split(';')
author_id_list = [author_id.strip() for author_id in author_id_list if author_id.strip() != '']
print(len(author_id_list)) 
```

```python
# Write a author_id_query_list which stores queries.
# Each query consists of only 50 users which is in the format of "AU-ID(author_id) OR" for all author_id in author_id_list
# For example, if author_id_list = ['123', '456', '789',...'182','298','291'], then author_id_query_list = ["AU-ID(123) OR AU-ID(456) OR AU-ID(789) OR...", "AU-ID(182) OR AU-ID(298) OR AU-ID(291) OR..."]

# Code here
author_id_query_list = []
author_id_query = ''
for i, author_id in enumerate(author_id_list):
    author_id_query += 'AU-ID({}) OR '.format(author_id)
    if (i + 1) % 50 == 0:
        author_id_query_list.append(author_id_query[:-4])
        author_id_query = ''
author_id_query_list.append(author_id_query[:-4])
print(len(author_id_query_list))
print(author_id_query_list[0])
```

```python
from pybliometrics.scopus import AuthorSearch
authors_search_list = []
authors_search_list.append(AuthorSearch(author_id_query_list[0],verbose=True))
```

```python
for author_id_query in author_id_query_list[1:]:
    authors_search_list.append(AuthorSearch(author_id_query,verbose=True))
```

```python
# Extract all the authors' information from the authors_search_list
# Store the information into a dataframe called authors_df

# Code here
authors_df = pd.DataFrame()
for authors_search in authors_search_list:
    # do not use the append method
    authors_df = pd.concat([authors_df, pd.DataFrame(authors_search.authors)], ignore_index=True)
authors_df = authors_df.drop_duplicates(subset=['eid'])
authors_df.head(5)
len(authors_df) # There should be 845 authors
```

```python
authors_df['au_id'] = authors_df['eid'].apply(lambda x: x.split('-')[2])
```

```python
# authors_df.to_csv('output/authors_2016_2017_2022.csv', index=False)
display(authors_df)
```

```python
# Parse a string like 'Rizzardini, Rocael Hernández (55918790500); Amado-Salvatierra, Hector R. (52663094800)' into 6 columns or more columns:
# 1. First author first name
# 2. First author last name
# 3. First author ID
# 4. Second author first name
# 5. Second author last name
# 6. Second author ID
# 7. Third author...


# Sample DataFrame with "Author full names" column
# data = {'Author full names': ['Rizzardini, Rocael Hernández (55918790500); Amado-Salvatierra, Hector R. (52663094800); Smith, John (1234567890)']}
# df = pd.DataFrame(data)

# Split the "Author full names" column by semicolon
df['Authors'] = df['Author full names'].str.split(';')

# Initialize lists to store author information
first_names = []
last_names = []
ids = []

# Iterate through the authors and extract information
for authors_list in df['Authors']:
    first_name_list = []
    last_name_list = []
    id_list = []
    for author in authors_list:
        parts = author.strip().split('(')
        name_parts = parts[0].split(',')
        last_name_list.append(name_parts[0].strip())
        first_name_list.append(name_parts[1].strip())
        id_list.append(parts[1].replace(')', '').strip())
    
    # Append author information to the respective lists
    first_names.append(first_name_list)
    last_names.append(last_name_list)
    ids.append(id_list)

# Determine the maximum number of authors
max_authors = max(len(authors) for authors in first_names)

# Create empty columns for author information
for i in range(1,max_authors+1):
    df[f'Author {i} First Name'] = [authors[i-1] if i-1 < len(authors) else None for authors in first_names]
    df[f'Author {i} Last Name'] = [authors[i-1] if i-1 < len(authors) else None for authors in last_names]
    df[f'Author {i} ID'] = [authors[i-1] if i-1 < len(authors) else None for authors in ids]

# Drop the intermediate columns
df.drop(columns=['Author full names', 'Authors'], inplace=True)

# Print the DataFrame
display(df)
```

```python
# Merge the author_df with the df DataFrame for Author i ID for i from 1 to max_authors
# Code here
for i in range(1, max_authors+1): 
    df = df.merge(authors_df[['au_id', 'affiliation','country']], left_on=f'Author {i} ID', right_on='au_id', how='left')
    df = df.rename(columns={'affiliation': f'Author {i} Affiliation', 'country': f'Author {i} Country'})
    df.drop(columns=['au_id'], inplace=True)
```
